```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def jogar_moeda(jogadas):
    cont_cara = 0
    cont_coroa = 0
    for i in range(jogadas):
        rand = random.randint(1,2)
        if rand == 1:
            cont_coroa += 1
        else:
            cont_cara +=1

    simulacao_dict = {'Lançamentos': [cont_cara, cont_coroa], 'Proporção': [(cont_cara/jogadas)*100, (cont_coroa/jogadas)*100]}

    tabela = pd.DataFrame(simulacao_dict, index=['Cara', 'Coroa'])
    print("A quantidade de arremessos e proporção são: ")
    print(tabela)

    sns.barplot(x=tabela.index, y=tabela['Lançamentos'])

jogar_moeda(500)
```
